```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import MNIST dataset from keras and load MNIST dataset
from keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
#print shapes of x_train, y_train, x_test and y_test
print(f'x_train.shape = {x_train.shape}')
print(f'y_train.shape = {y_train.shape}')
print(f'x_test.shape = {x_test.shape}')
print(f'y_test.shape = {y_test.shape}')
```

```python
#find unique values in y_train and y_test which represent all unqiue classes in MNIST dataset
unique_in_train = np.unique(y_train)
print(f'unique_in y_train = {unique_in_train}')
unique_in_test = np.unique(y_test)
print(f'unique in y_test = {unique_in_test}')
```

```python
#reshape images in x_train and x_test into vectors
# change dtype of values in x_train and x_test to float32
x_train = x_train.reshape(x_train.shape[0], 28 * 28).astype('float32')
x_test = x_test.reshape(x_test.shape[0], 28 * 28).astype('float32')
print(f'x_train.shape = {x_train.shape}, x_test.shape = {x_test.shape}')
print(f'x_train[0] = {x_train[0]}')
```

```python
# normalize x_train and x_test in range of [0; 1]
x_train /= 255
x_test /= 255
```

```python
#import kNN Classifier from scikit-learn
from sklearn.neighbors import KNeighborsClassifier as kNNClassifier
from sklearn.neighbors import KNeighborsRegressor as kNNRegressor
```

```python
#initialize kNN Classification model
algorithm = 'brute'
metric = 'euclidean'
k = 3      
model = kNNClassifier(n_neighbors = k, algorithm = algorithm, metric = metric)
```

```python
#train kNN Classification model
model.fit(x_train, y_train)
```

```python
#predict with kNN Classification model
y_test[3] = model.predict(x_test[3].reshape(1, -1))
print(y_test[3])
```

```python
#plot selected image from x_test and its respective original label from y_test
plt.figure(1, figsize=(4, 4))
plt.imshow(x_test[3].reshape(28, 28), cmap='gray')
plt.suptitle(f'predicted x_test[3]\nk = {k},  y_test[3] = {y_test[3]}')
plt.show()
```

```python
#prob test with k = 1, 2
#algorithm = 'brute', new_metric = 'euclidean'
new_algorithm = 'brute'
new_metric = ['euclidean', 'cosine', 'minkowski', 'chebyshev']
print(f'new_metric = {new_metric[0]}')
lst_accuracy_train = []
lst_accuracy_test = []
lst_accuracy_train.append([])
print(lst_accuracy_train[0])
lst_accuracy_train.append(2)
print(lst_accuracy_train)
plt.figure(1, figsize=(4, 4))
plt.title("Accuracy of kNN models")
plt.xlabel("k")
plt.ylabel("accuracy")
lst_accuracy_train.append([])
lst_accuracy_test.append([])
for k in range(1, 3):
    print(f'k = {k}')
    # initialize kNN Classification model with new k value
    model = kNNClassifier(n_neighbors = k, algorithm = new_algorithm, metric = new_metric[0])
    
    # fit model
    model.fit(x_train, y_train)
    
    # predict training subset
    y_train_pred = model.predict(x_train)
    
    # predict testing subset
    y_test_pred = model.predict(x_test)
    
    # calculate model's accuracy for training subset
    accuracy_train = model.score(x_train, y_train)

    # save accuracy_train
    lst_accuracy_train[0].append(accuracy_train)
    
    # calculate model's accuracy for testing subset, save the value
    accuracy_test = model.score(x_test, y_test)

    # save accuracy_test
    lst_accuracy_test[0].append(accuracy_test)
    
print(f'accuracy_train = {accuracy_train}, accuracy_test = {accuracy_test} ')
print(f'lst_accuracy_test = {lst_accuracy_test}')       


plt.plot(np.arange(1, 3), lst_accuracy_train[0], "8-b", label='euclidean, train')
plt.plot(np.arange(1, 3), lst_accuracy_test[0], "^--b", label='euclidean, test')
plt.legend(loc='upper right')
plt.show()
```
